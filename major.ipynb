```python
import tensorflow as tf
from tensorflow.keras import models,layers
import os
import cv2

import pickle
import numpy as np
import time
```

```python
gpu_device = tf.config.experimental.list_physical_devices('GPU')
print(gpu_device)
print(len(gpu_device))
tf.config.experimental.set_memory_growth(gpu_device[0],True)
```

```python
height = 360
width = 480
```

```python
model = models.Sequential()
model.add(layers.Conv2D(filters=64,kernel_size=9, padding = 'same', activation='relu', input_shape = (height,width,3)))
model.add(layers.LeakyReLU(alpha = 0.1))
model.add(layers.Conv2D(filters=16,kernel_size=1, padding = 'same', activation='relu'))
model.add(layers.Conv2D(filters=32,kernel_size=7, padding = 'same', activation='relu'))
model.add(layers.LeakyReLU(alpha = 0.1))
model.add(layers.Conv2D(filters=8,kernel_size=1, padding = 'same', activation='relu'))
model.add(layers.Conv2D(filters=16,kernel_size=3, padding = 'same', activation='relu'))
model.add(layers.LeakyReLU(alpha = 0.1))
model.add(layers.Conv2D(filters=8,kernel_size=1, padding = 'same', activation='relu'))
model.add(layers.Conv2D(filters=3,kernel_size=5, padding = 'same', activation='relu'))
model.compile(optimizer='adam', metrics=['accuracy'], loss = 'mean_squared_error')
```

```python
model.summary()
```

```python
X = []
y = []
```

```python
X_path = 'D:\\major_proj\\compressed'
y_path = 'D:\\major_proj\\resized_data'
```

```python
for i in os.listdir(X_path)[:500]:
    img_name = i.split('.')[0]
    X.append(cv2.imread(os.path.join(X_path,(img_name+'.jpeg'))))
    y.append(cv2.imread(os.path.join(y_path,(img_name+'.jpg'))))
```

```python
#1,360,480,3
ssim_arr = []
psnr_arr = []
new_psnr_arr = []
new_ssim_arr = []
for i in os.listdir(X_path)[2000:2100]:
    img_name = i.split('.')[0]
    actual_img = cv2.imread(os.path.join(y_path,(img_name + '.jpg')))
    curr_img = cv2.imread(os.path.join(X_path,(img_name + '.jpeg')))
    curr_img = curr_img.reshape(1,360,480,3)
    curr_img = curr_img / 255
    curr_out = model.predict(curr_img)
    curr_out = curr_out * 255
    psnr_arr.append(tf.cast(tf.image.psnr(curr_out[0].astype('float32'), actual_img.astype('float32'), max_val=255), dtype='float'))
    ssim_arr.append(tf.cast(tf.image.ssim(curr_out[0].astype('float32'), actual_img.astype('float32'), max_val=255), dtype='float'))
    
    
```

```python
new_psnr_arr = []
new_ssim_arr = []

for i in os.listdir(X_path)[2000:2100]:
    img_name = i.split('.')[0]
    actual_img = cv2.imread(os.path.join(y_path,(img_name + '.jpg')))
    curr_img = cv2.imread(os.path.join(X_path,(img_name + '.jpeg')))
    new_psnr_arr.append(tf.cast(tf.image.psnr(curr_img.astype('float32'), actual_img.astype('float32'), max_val=255), dtype='float'))
    new_ssim_arr.append(tf.cast(tf.image.ssim(curr_img.astype('float32'), actual_img.astype('float32'), max_val=255), dtype='float'))
```
